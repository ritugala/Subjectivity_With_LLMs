```python
columns_to_check = data.columns[4:18]
for col in columns_to_check:
    data[col] = data[col].str.lower()
data = data[data[columns_to_check].isin(['yes', 'no']).all(axis=1)]
```

```python
columns = ['Filler Words', 'Stuttering', 'Concise', 'Repetition', 'Rambling', 
           'External References', 'Complicated Terms', 'Sarcasm', 
           'Rhetorical Questions', 'Direct Answer', 'Logical Coherence', 
           'Hedging Language', 'Assertive Language', 'Evasive Language']

# Binarizing the columns: 1 for 'yes', 0 for 'no'
for col in columns:
    data[col] = data[col].map({'yes': 1, 'no': 0})
```

```python
import pandas as pd
import os

# Ensure New_Features_Binary is a 14-character binary string
data['New_Features_Binary'] = data['New_Features_Binary'].apply(lambda x: x.zfill(14))

# Base directory for CongressionalHearingFolds
base_dir = '/Users/meghanaarajeev/Documents/ANLP/Subjectivity_With_LLMs/subjective_discourse/data/gold/CongressionalHearingFolds'

# Folds
folds = ['fold0', 'fold1', 'fold2', 'fold3']

# Iterate over each fold
for fold in folds:
    train_csv_path = os.path.join(base_dir, fold, 'train.tsv')
    train_df = pd.read_csv(train_csv_path, sep="\t")

    # Convert 'qa_index_digits' to int for consistent comparison
    train_df['qa_index_digits'] = train_df['qa_index_digits'].astype(int)

    # Calculate the difference between 'Index' values in 'data' and 'qa_index_digits' in train_df
    missing_indices = set(data['Index']) - set(train_df['qa_index_digits'])

    # Print the missing 'Index' values for each fold
    print(f"Missing Index values in {fold}: {missing_indices}")


    
```
